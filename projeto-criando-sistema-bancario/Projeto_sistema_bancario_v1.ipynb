## Projeto Sistema Bancário - v1

```python
#O objetivo é implementar três operações essenciais: depósito, saque e extrato. Todos os depositos e saques devem ser armazaenados em uma variável e exibidos na operação de extrato.

#Pré requisitos:
#Deposito -> ser positivo
#Saque -> maior que o limite, valor positivo e ser menor que o saldo. 
#Extrato -> exibir todas as operações realizadas.
```

## Solução 1

```python
menu = """MENU

[d] Depositar
[s] Saque
[e] Extrato
[q] Sair

"""

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:
    opcao = input(menu)
    
    if opcao =="d":
        valor = float(input("Informe o valor do depósito: "))
        if valor > 0:
            saldo += valor #adicionando o valor do depósito no saldo
            extrato += f"Depósito: R$ {valor: .2f}\n" #salvando o valor na variável extrato na formatação correta
            print("\nDepósito realizado com sucesso!")
        else:
            print("\nOperação inválida, valor não permitido. Tente novamente!")          
               
    elif opcao == "s":
        
        if numero_saques <= LIMITE_SAQUES: #verifica o limite de saques diários
            valor = float(input("Qual o valor do saque? "))
                       
            if valor < saldo: #verifica o saldo  
                if valor > 0 and valor <= 500: #verifica se é positivo e inferior ao limite de saque
                    saldo -= valor
                    extrato += f"Saque: R$ {valor: .2f}\n"
                    numero_saques += 1 #adiciona o saque na variável para verificar a quantidade de saques realizados no dia
                    print("\nSaque realizado com sucesso!")
                else:
                    print("\nOperação inválida, excedeu o valor limite de saque.")
                
            else:
                print("\nSaldo insuficiente para saque.")
        else:
            print("Limite de saques diários excedido, tente novamente amanhã.")
            
    elif opcao == "e":
        print("\n############# EXTRATO #############")
        print("Não fora realizadas operações." if not extrato else extrato)
        print(f"\nSaldo: R$ {saldo:.2f}")
        print("###################################")
    
    elif opcao =="q":
        break
        
    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```

## Solução 2

```python
#Menu
menu = """MENU

[d] Depositar
[s] Saque
[e] Extrato
[q] Sair

"""
#Variáveis
saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:
    opcao = input(menu)
    
    if opcao =="d":
        valor = float(input("Qual o valor do depósito? "))
        if valor > 0:
            saldo += valor
            extrato += f"Depósito: R$ {valor: .2f}\n"
            print("\nDepósito realizado com sucesso!")
        else:
            print("\nOperação inválida, valor não permitido. Tente novamente!")          
               
    elif opcao == "s":
        
        valor = float(input("Qual o valor do saque? "))
        
        excedeu_saldo = valor >= saldo
        excedeu_limite = valor >= limite
        excedeu_saques = numero_saques >= LIMITE_SAQUES
        
        if excedeu_saldo:
            print("\nSaldo insuficiente para completar a operação.")
        
        elif excedeu_limite:
            print("\nO valor do saque excede o limite de R$ 500,00 por saque.")
            
        elif excedeu_saques:
            print("\nLimite de saques diários excedido, tente novamente amanhã!")
            
        elif valor > 0:
            saldo -= valor
            extrato += f"Saque: R$ {valor: .2f}\n"
            numero_saques += 1
            print("Saque realizado com sucesso!")
            
        else:
            print("Operação inválida!")
            
    elif opcao == "e":
        print("\n############# EXTRATO #############")
        print("Não fora realizadas operações." if not extrato else extrato)
        print(f"\nSaldo: R$ {saldo:.2f}")
        print("###################################")
    
    elif opcao =="q":
        break
        
    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```
